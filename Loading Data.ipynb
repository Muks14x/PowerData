```python
df_loaded = pd.read_pickle('clean_data.pkl')
df_loaded['W'] = df_loaded[0]
del(df_loaded[0])
```

```python
tempDf
```

```python
lastDate = list(df_loaded.tail(1)['DateTime'])[0]
lastDay = lastDate.day
lastMon = lastDate.month

train_idx = tempDf[(tempDf['Month'] == lastMon) & (tempDf['Day'] == lastDay)].index[0] + 1
tempTrain = tempDf[: train_idx]
tempTest = tempDf[train_idx: ]
```

```python
tempTest
```

```python
# weekday - only one week's vector
weekday = []
for i in [24 * [i] for i in [0, 0, 0, 1, 1, 0, 0]]:
    weekday += i
X_weekday = weekday * (len(df_loaded)//7//24) + weekday[:(len(df_loaded)%(7 * 24) * 24)]
X_weekday_test = weekday * (len(tempTest)//7) + weekday[:((len(tempTest)%7) * 24)]
```

```python
len(X_weekday), len(X_weekday_test)
```

```python
len(X_weekday)
```

```python
act_h = list(tempTrain['act_h'])
act_l = list(tempTrain['act_l'])
act_h_test = list(tempTest['act_h'])
act_l_test = list(tempTest['act_l'])
```

```python
X_hT = []
X_lT = []
X_hT_test = []
X_lT_test = []

for i in range(len(tempTrain)):
    X_hT += [act_h[i]] * 24
    X_lT += [act_l[i]] * 24
    
for i in range(len(tempTest)):
    X_hT_test += [act_h_test[i]] * 24
    X_lT_test += [act_l_test[i]] * 24
```

```python
len(X_hT), len(X_lT), len(X_hT_test), len(X_lT_test)
```

```python
df_loaded['HiT'] = X_hT
df_loaded['LoT'] = X_lT
df_loaded['isWeekend'] = X_weekday
```

```python
df_loaded
```

```python
df_test = pd.DataFrame()
df_test['HiT'] = X_hT_test
df_test['LoT'] = X_lT_test
df_test['isWeekend'] = X_weekday_test
```
