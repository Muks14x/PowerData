```python
def mmddyyyy(x):
    if type(x) == str and len(x) == 10:
        return x[3:6]+x[0:3]+x[6:]
    else:
        return x
```

```python
full_data['DateTime'] = pd.to_datetime(full_data.Date.apply(mmddyyyy) + ' ' + full_data.Time)
```
